```
import numpy as np
import pandas as pd
```

```
from kmodes.kprototypes import KPrototypes
```

```
marketing_df = pd.read_csv('https://raw.githubusercontent.com/srivatsan88/YouTubeLI/master/dataset/marketing_cva_f.csv')
```

```
marketing_df.columns
```

```
marketing_df.head()
```

```
marketing_df = marketing_df.drop(['Customer'], axis = 1)
```

```
marketing_df
```

```
mark_array = marketing_df.values
```

```
mark_array[:,1] = mark_array[:,1].astype(float)
mark_array[:,3] = mark_array[:,3].astype(float)
mark_array[:,5] = mark_array[:,5].astype(float)
mark_array[:,6] = mark_array[:,6].astype(float)
mark_array[:,7] = mark_array[:,7].astype(float)
mark_array[:,8] = mark_array[:,8].astype(float)
mark_array[:,10] = mark_array[:,10].astype(float)
```

```
mark_array
```

```
kpt =KPrototypes(n_clusters=5, verbose=1,  max_iter = 20)
clusters = kpt.fit_predict(mark_array, categorical = [0,2,4,9])
```

```
print(kpt.cluster_centroids_)
```

```
cluster_list = []
for c in clusters:
  cluster_list.append(c)
```

```
marketing_df["predicted_cluster"] = cluster_list
```

```
marketing_df[marketing_df["predicted_cluster"] == 1]
```

```
marketing_df[marketing_df["predicted_cluster"] == 0]
```

```
marketing_df[marketing_df["predicted_cluster"] == 2]
```
